```python
import pandas as pd
from sklearn import tree

df = pd.read_csv("cervejas.csv")
df
```

```python
features = ["copo", "espuma", "cor","temperatura"]
target = ["classe"]

x = df[features]
y = df[target]

x = x.replace({
    "copo": {
        "Snifter": 1,
        "Pint": 2,
        "Tulipa": 3,
        "Weizen": 4,
        "Caneca": 5,
        "Taça": 6,
        "Outro": 7
    },
    "espuma": {
        "Baixa": 1,
        "Média": 2,
        "Alta": 3
    },
    "cor": {
        "Clara": 1,
        "Âmbar": 2,
        "Escura": 3
    },
})

```

```python
model = tree.DecisionTreeClassifier()
model.fit(x,y)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=400)
tree.plot_tree(model, feature_names=features, class_names=model.classes_, filled=True)
```
